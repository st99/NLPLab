## Evaluate unigram tagger using brown corpus.

```
from nltk.corpus import brown
from nltk.tag import UnigramTagger
```

Evaluating unigram tagger

```
tagged_sents = brown.tagged_sents(categories="news")
uni_tagger = UnigramTagger(tagged_sents)
uni_tagger.tag(tagged_sents[2011])
print(f"Evaluation of unigram tagger:: {round(uni_tagger.evaluate(tagged_sents), 4)}")
```

Training the unigram tagger

```
tagged_sents = brown.tagged_sents(categories="news",tagset="universal")
i = int(len(tagged_sents)*0.90)
training_data = tagged_sents[:i]
test_data = tagged_sents[i:]
uni_tagger = UnigramTagger(training_data)
```

Testing the unigram tagger

```
tagged_test_data = uni_tagger.tag_sents([[token for token, tag in sent] for sent in test_data])
print("Actual data:: ",test_data[0])
print("Test data:: ",tagged_test_data[0])
print(f"Evaluation of unigram tagger using test data:: {round(uni_tagger.evaluate(test_data), 4)}")
```

Measuring precision, recall, f1-score and support

```
actual = [str(tag) for sent in test_data for token, tag in sent]
pred = [str(tag) for sent in tagged_test_data for token, tag in sent]

from sklearn import metrics
print(metrics.classification_report(actual, pred))
```

## Combining the Taggers

Training the combined tagger

```
tagged_sents = brown.tagged_sents(categories="news",tagset="universal")
i = int(len(tagged_sents)*0.90)
training_data = tagged_sents[:i]
test_data = tagged_sents[i:]
```

```
def backoff_tagger(train_sents, tagger_classes, backoff=None):
    for cls in tagger_classes:
        backoff = cls(train_sents, backoff=backoff)
    return backoff
    
from nltk.tag import DefaultTagger

back_tagger = DefaultTagger('NN')
combined_tagger = backoff_tagger(training_data, [UnigramTagger, BigramTagger, TrigramTagger], backoff=back_tagger)
```

Testing the combined tagger

```
tagged_test_data = combined_tagger.tag_sents([[token for token, tag in sent] for sent in test_data])
print("Actual data:: ",test_data[0])
print("Test data:: ",tagged_test_data[0])
print(f"Evaluation of combine tagger using test data:: {round(combined_tagger.evaluate(test_data), 4)}")
```

Measuring precision, recall, f1-score and support

```
actual = [str(tag) for sent in test_data for token, tag in sent]
pred = [str(tag) for sent in tagged_test_data for token, tag in sent]

print(metrics.classification_report(actual, pred, zero_division=1))
```

## Brill Tagger

Defining the Brill trainer


```
from nltk.tag import brill, brill_trainer

def train_brill_tagger(initial_tagger, train_sents, **kwargs):
    templates = [
                 brill.Template(brill.Pos([-1])),
                 brill.Template(brill.Pos([1])),
                 brill.Template(brill.Pos([-2])),
                 brill.Template(brill.Pos([2])),
                 brill.Template(brill.Pos([-2, -1])),
                 brill.Template(brill.Pos([1, 2])),
                 brill.Template(brill.Pos([-3, -2, -1])),
                 brill.Template(brill.Pos([1, 2, 3])),
                 brill.Template(brill.Pos([-1]), brill.Pos([1])),
                 brill.Template(brill.Word([-1])),
                 brill.Template(brill.Word([1])),
                 brill.Template(brill.Word([-2])),
                 brill.Template(brill.Word([2])),
                 brill.Template(brill.Word([-2, -1])),
                 brill.Template(brill.Word([1, 2])),
                 brill.Template(brill.Word([-3, -2, -1])),
                 brill.Template(brill.Word([1, 2, 3])),
                 brill.Template(brill.Word([-1]), brill.Word([1])),
    ]
    trainer = brill_trainer.BrillTaggerTrainer(initial_tagger, templates, deterministic=True)
    return trainer.train(train_sents, **kwargs)
```

Training Brill and combined tagger


```
from nltk.tag import brill, brill_trainer
from nltk.tag import DefaultTagger
from nltk.corpus import brown

tagged_sents = brown.tagged_sents(categories="news",tagset="universal")
i = int(len(tagged_sents)*0.90)
training_data = tagged_sents[:i]
test_data = tagged_sents[i:]

#Brill Tagger
brill_tagger = train_brill_tagger(combined_tagger, training_data)

#Combined Tagger
back_tagger = DefaultTagger('NN')
combined_tagger = backoff_tagger(training_data, [UnigramTagger, BigramTagger, TrigramTagger], backoff=back_tagger)
```

Testing both the taggers

```
tagged_test_data = brill_tagger.tag_sents([[token for token, tag in sent] for sent in test_data])
print("Actual data:: ",test_data[0])
print("Test data:: ",tagged_test_data[0])
print(f"Evaluation of Brill Tagger::  {round(brill_tagger.evaluate(test_data), 4)}")
```

Measuring precision, recall, f1-score and support

```
actual = [str(tag) for sent in test_data for token, tag in sent]
pred = [str(tag) for sent in tagged_test_data for token, tag in sent]

print(metrics.classification_report(actual, pred, zero_division=1))
```

Comparing accuracies

```
combined_accuracy = combined_tagger.evaluate(test_data)
print("Accuracy of Combined Tagger:: ", combined_accuracy)

brill_accuracy = brill_tagger.evaluate(test_data)
print("Accuracy of Brill Tagger:: ", brill_accuracy)
```
