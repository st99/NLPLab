```
nltk.download('names')
from nltk.corpus import names

print("# Male Names : ", len(names.words('male.txt')))
print("# Female Names : ", len(names.words('female.txt')))
```

```
mname = names.words('male.txt')
fname = names.words('female.txt')

print("Male Names:: ")
print(mname[:10])
print("Female Names:: ")
print(fname[:10])
```

6.

```
male_name = [(name, 'male') for name in mname]
female_name = [(name, 'female') for name in fname]
combined_name = male_name + female_name
total_len = len(male_name)+len(female_name)

import random
for _ in range(15):
    i = random.randrange(total_len) ##total number of names
    print(combined_name[i])
```
